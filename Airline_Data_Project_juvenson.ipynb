Safety Metric Calculation : 



```python
# first lets determine the total of the overall Injuries
df_20['Total.Injuries'] = df_20[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum(axis=1)
```

```python
df_20['Fatality.Rate'] = df_20['Total.Fatal.Injuries'] / df['Total.Injuries']
```
